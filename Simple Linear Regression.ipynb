```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
print(df)
```

```python
df.head()
```

```python
x=df.area
x
```

```python
y=df.price
y
```

```python
col=np.array([10,80,39,46,65])
plt.scatter(x,y,marker='*',s=150,c=col,cmap="plasma")
plt.colorbar()
plt.xlabel('Area',c='g',size=20)
plt.ylabel('Price',c='r',size=20)
plt.title("Scatter Plot",c='gold',size=15)
```

### By looking at the graph i can understand the distribution is suitable for linear regression model

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)    #i fit my model,fitting data means = training the linear regression model using available data points

#now this linear regression model is ready to predict the prices
```

```python


plt.scatter(df.area,df.price,marker='*',s=150,c='r')

plt.xlabel('Area',c='g',size=20)
plt.ylabel('Price',c='r',size=20)
plt.title("Scatter Plot",c='gold',size=15)

plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

### Predicting the prices--

#### (1) Predict price of a home with area = 3300 sqr ft

```python
reg.predict([[3300]])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('salary_data.csv')
df
```

```python
x=df.YearsExperience
y=df.Salary
```

```python
x=df[['YearsExperience']]
y=df['Salary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model.predict([[12]])
```

```python
model.predict(x_test)
```
